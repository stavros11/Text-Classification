```python
import collections

def find_tree_leaves(root):
    nodes = collections.deque([root])
    leaves = []
    while nodes:
        node = nodes.popleft()
        if hasattr(node, "children"):
            for child in node.children:
                nodes.append(child)
        else:
            leaves.append(node)
    return leaves
```

### How to find where reviews are located in the JS code

```python
def find_reviews(root, target):
    traces = [""]
    stack = [root]
    while stack:
        node = stack.pop()
        trace = traces.pop()
        if isinstance(node, dict):
            for k, v in node.items():
                traces.append("/".join([trace, k]))
                stack.append(v)
        elif isinstance(node, list):
            for i, v in enumerate(node):
                traces.append("/".join([trace, str(i)]))
                stack.append(v)
        elif isinstance(node, str):
            if target in node:
                return trace
    return None
```
